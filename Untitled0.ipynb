```python
import pandas as pd
import plotly.express as px

# URLs dos datasets
orders_url = 'https://raw.githubusercontent.com/olist/work-at-olist-data/master/datasets/olist_orders_dataset.csv'
customers_url = 'https://raw.githubusercontent.com/olist/work-at-olist-data/master/datasets/olist_customers_dataset.csv'
order_items_url = 'https://raw.githubusercontent.com/olist/work-at-olist-data/master/datasets/olist_order_items_dataset.csv'
products_url = 'https://raw.githubusercontent.com/olist/work-at-olist-data/master/datasets/olist_products_dataset.csv'
payments_url = 'https://raw.githubusercontent.com/olist/work-at-olist-data/master/datasets/olist_order_payments_dataset.csv'
reviews_url = 'https://raw.githubusercontent.com/olist/work-at-olist-data/master/datasets/olist_order_reviews_dataset.csv'

# Carregar os datasets
orders = pd.read_csv(orders_url)
customers = pd.read_csv(customers_url)
order_items = pd.read_csv(order_items_url)
products = pd.read_csv(products_url)
payments = pd.read_csv(payments_url)
reviews = pd.read_csv(reviews_url)

# Contagem dos m√©todos de pagamento
payment_counts = payments['payment_type'].value_counts().reset_index()
payment_counts.columns = ['payment_type', 'count']

# Plotar o gr√°fico de pizza
fig = px.pie(payment_counts, values='count', names='payment_type', title='Payment Method Distribution')
fig.show()

```

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])

orders['shipping_time'] = (orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']).dt.days

print("Mean shipping time:", orders['shipping_time'].mean())
print("Median shipping time:", orders['shipping_time'].median())
print("Standard deviation of shipping times:", orders['shipping_time'].std())

# Visualization of shipping times
plt.figure(figsize=(10, 6))
sns.boxplot(x=orders['shipping_time'])
plt.title('Boxplot of Shipping Times')
plt.xlabel('Shipping Time (days)')
plt.show()
```

```python
reviews['review_score'] = reviews['review_score'].astype(int)
print("Mean satisfaction rating:", reviews['review_score'].mean())
print("Median satisfaction rating:", reviews['review_score'].median())

satisfied_customers = reviews[reviews['review_score'] >= 4].shape[0]
dissatisfied_customers = reviews[reviews['review_score'] < 4].shape[0]
total_customers = reviews.shape[0]

print("Proportion of satisfied customers:", satisfied_customers / total_customers)
print("Proportion of dissatisfied customers:", dissatisfied_customers / total_customers)

px.pie(values=[satisfied_customers, dissatisfied_customers], names=['Satisfied', 'Dissatisfied'], title='Customer Satisfaction Distribution').show()
```

```python
# Define null and alternative hypotheses
# H0: There is no difference in customer satisfaction between orders with standard shipping and orders with express shipping.
# H1: Orders with express shipping have higher customer satisfaction compared to orders with standard shipping.

# Create two groups based on shipping time (assuming standard shipping is more than the median shipping time)
median_shipping_time = orders['shipping_time'].median()
standard_shipping = reviews[orders['shipping_time'] > median_shipping_time]['review_score']
express_shipping = reviews[orders['shipping_time'] <= median_shipping_time]['review_score']

# Conduct a t-test
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(standard_shipping, express_shipping)
print("t-statistic:", t_stat)
print("p-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant difference in satisfaction ratings between standard and express shipping methods.")
else:
    print("Fail to reject the null hypothesis. No significant difference in satisfaction ratings between standard and express shipping methods.")
```

```python
# Visualization of hypothesis testing result
data = {'Group': ['Standard Shipping', 'Express Shipping'],
        'Mean Satisfaction Rating': [standard_shipping.mean(), express_shipping.mean()]}
df = pd.DataFrame(data)
px.bar(df, x='Group', y='Mean Satisfaction Rating', title='Customer Satisfaction: Standard vs Express Shipping').show()
```
